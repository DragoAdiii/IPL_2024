```python
import pandas as pd
from google.colab import files
from io import StringIO
import matplotlib.pyplot as plt

# Upload file
uploaded = files.upload()

# Print the file names to check
print("Uploaded files:", uploaded.keys())

# Load the uploaded file into a DataFrame
file_name = list(uploaded.keys())[0]  # Get the name of the uploaded file
df = pd.read_csv(StringIO(uploaded[file_name].decode('utf-8')))

# Display the first few rows to understand the structure
print(df.head())
```

```python
print(df.columns)
```

```python
# 1. Pie Chart - Distribution of wins by team
plt.figure(figsize=(10, 6))
df['winner'].value_counts().plot(kind='pie', autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired(range(len(df['winner'].unique()))))
plt.title('Distribution of Wins by Teams')
plt.ylabel('')
plt.show()
```

```python

# 2. Histogram - Scores of teams in the first innings
plt.figure(figsize=(10, 6))
plt.hist(df['first_score'], bins=20, color='skyblue', edgecolor='black')
plt.title('Histogram of First Innings Scores')
plt.xlabel('Score')
plt.ylabel('Frequency')
plt.show()
```

```python

# 3. Bar Plot - Number of matches won by each team
plt.figure(figsize=(12, 8))
df['winner'].value_counts().plot(kind='bar', color='orange')
plt.title('Number of Matches Won by Each Team')
plt.xlabel('Teams')
plt.ylabel('Number of Wins')
plt.xticks(rotation=45)
plt.show()
```

```python

# 4. Line Plot - Scores over matches (combined first and second innings)
df['total_score'] = df['first_score'] + df['second_score']
plt.figure(figsize=(10, 6))
df.groupby('date')['total_score'].mean().plot(kind='line', marker='o', color='green')
plt.title('Average Total Score Over Matches')
plt.xlabel('Date')
plt.ylabel('Average Total Score')
plt.grid(True)
plt.xticks(rotation=45)
plt.show()
```
